```python
import pandas as pd

#TODO: also add column names to each file

prev_ticker = None

def iter_chunk_by_id(file):
    csv_reader = pd.read_csv(file, iterator=True, chunksize=1, header=None)
    first_chunk = csv_reader.get_chunk()
    id = first_chunk.iloc[0,5]
    chunk = pd.DataFrame(first_chunk)
    for l in csv_reader:
        if id == l.iloc[0, 5]:
            id = l.iloc[0, 5]
            chunk = chunk.append(l)
            continue
        id = l.iloc[0, 5]
        yield chunk
        chunk = pd.DataFrame(l)
    yield chunk

chunk_iter = iter_chunk_by_id("iris.csv")

i = 0
cols = None
for chunk in chunk_iter:
    print(chunk)
    print("_____")

    if i == 0:
        cols = chunk.loc[0, 1:].values.tolist()
    else:
        del chunk[0]
        chunk.columns = cols
        chunk.to_csv('data/'+str(i)+'.csv')
    i += 1
```
